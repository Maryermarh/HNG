```python
#importing necessary libraries for the analysis
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#reading the dataset
df = pd.read_csv('SampleSuperstore.csv')
```

```python
df["Sub-Category"].unique()  #getting the different types of items
```

```python
df.groupby("Category")["Sales"].sum() #getting the sum of all category of items sold 
```

```python
df.groupby("Category")["Profit"].sum() #getting profit value for all the category of items 
```

```python
top_selling_products = df.groupby("Category")["Sales"].sum().sort_values(ascending=False)
print(top_selling_products)  # Display the top selling products
```

```python
# visualize the top selling categories
plt.figure(figsize=(10, 6))
sns.barplot(x=top_selling_products.values, y=top_selling_products.index, palette="viridis")

plt.xlabel("Total Sales")
plt.ylabel("Category")
plt.title("Top Selling Product Categories")
plt.gca().invert_yaxis()  # To show the highest value at the top

plt.show()
```

```python
top_selling_sub_products = df.groupby("Sub-Category")["Sales"].sum().sort_values(ascending=False)
print(top_selling_sub_products)  # Display the top-selling items 
```

```python
# Category with top profit
plt.figure(figsize=(10, 6))
sns.barplot(x=top_selling_sub_products.values, y=top_selling_sub_products.index, palette="viridis")

plt.xlabel("Profit")
plt.ylabel("sub products")
plt.title("Top Selling S
ub Product Categories")

plt.show()
```

```python
# Category with top profit
plt.figure(figsize=(10, 6))
sns.barplot(x=top_selling_sub_products.values, y=top_selling_sub_products.index, palette="viridis")

plt.xlabel("Profit")
plt.ylabel("Sub-Category")
plt.title("Top Selling Product Categories")
plt.gca().invert_yaxis()  # To show the highest value at the top

plt.show()
```

```python
top_selling_products = df.groupby("Category")["Profit"].sum().sort_values(ascending=False)
print(top_selling_products.head(10))  # Display the top 10 selling products
```

```python
# Category with top profit
plt.figure(figsize=(10, 6))
sns.barplot(x=top_selling_products.values, y=top_selling_products.index, palette="viridis")

plt.xlabel("Profit")
plt.ylabel("Category")
plt.title("Profitability of Products by Category")
plt.gca().invert_yaxis()  # To show the highest value at the top

plt.show()
```
